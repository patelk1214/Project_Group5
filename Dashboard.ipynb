```python
import plotly.express as px
import panel as pn
from panel.interact import interact
from panel import widgets
import pandas as pd
import os
from pathlib import Path
import os
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import hvplot as hv
import hvplot.pandas
pn.extension('plotly')
```

```python
csvpath = Path("option_data/Final_option_dataframe.csv")
options_all = pd.read_csv(csvpath)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

options_all.drop(columns=['Unnamed: 0'], inplace=True)
options_all.head()
```

```python
is_march = options_all['exp_date']==20210319
march_maturities = options_all[is_march]
march_maturities.loc[march_maturities['symbol']=='DAL'].head(10)
```

```python
def get_df(symbol): 
    df = march_maturities.loc[march_maturities['symbol'] == symbol]
    return df

march_symbol = march_maturities.symbol.unique()

def plot_smile(symbol):
    df = get_df(symbol)

    
    return px.line(df,
        x = "strikePrice",
        y = "implied_vol",
        title="Volatility Smile of Call Options")
```

```python
def get_df1(symbol): 
    df1 = march_maturities.loc[march_maturities['symbol'] == symbol]
    return df1

march_symbol = march_maturities.symbol.unique()

def plot_smile1(symbol):

    df1 = get_df1(symbol)
    
    return px.line(df1,
        x = "strikePrice",
        y = "implied_vol.1",
        title="Volatility Smile of Put Options")
```

```python
def get_df3(symbol): 
    df3 = march_maturities.loc[march_maturities['symbol'] == symbol]
    return df3

march_symbol = march_maturities.symbol.unique()

def option_chain(symbol):
    
    df3 = get_df3(symbol)
    return df3.hvplot.table(columns=['strikePrice', 'op_Int','volume','vega','implied_vol','delta', 'bid','ask','side','side.1','ask.1','bid.1','delta.1','implied_vol.1','vega.1','volume.1','op_Int.1', 'Underlying_price', 'hist_vol'], 
                            sortable=True, selectable=True, width=1200, height=800)
```

```python
is_april = options_all['exp_date']==20210416
april_maturities = options_all[is_april]
# april_maturities.head(3)
```

```python
def get_df4(symbol): 
    df4 = april_maturities.loc[april_maturities['symbol'] == symbol]
    return df4

april_symbol = april_maturities.symbol.unique()

def plot_smile3(symbol):
    df4 = get_df4(symbol)
    
    return px.line(df4,
        x = "strikePrice",
        y = "implied_vol",
        title="Volatility Smile of Call Options")
```

```python
def get_df5(symbol): 
    df5 = april_maturities.loc[april_maturities['symbol'] == symbol]
    return df5

april_symbol = april_maturities.symbol.unique()

def plot_smile4(symbol):
    df5 = get_df5(symbol)
    
    return px.line(df5,
        x = "strikePrice",
        y = "implied_vol.1",
        title="Volatility Smile of Put Options")
```

```python
def get_df6(symbol): 
    df6 = april_maturities.loc[april_maturities['symbol'] == symbol]
    return df6

april_symbol = april_maturities.symbol.unique()

def option_chain1(symbol):
    df6 = get_df6(symbol)

    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', -1)

    return df6.hvplot.table(columns=['strikePrice', 'op_Int','volume','vega','implied_vol','delta', 'bid','ask','side','side.1','ask.1','bid.1','delta.1','implied_vol.1','vega.1','volume.1','op_Int.1', 'Underlying_price', 'hist_vol'], 
                            sortable=True, selectable=True,  width=1200, height=800)
```

```python
row = pn.Row(interact(plot_smile, symbol = march_symbol),interact(plot_smile1, symbol = march_symbol))

# march_demo_df = march_maturities.loc[march_maturities['symbol']=='AAPL']
# march_demo_df1 = march_maturities.loc[march_maturities['symbol']=='SPY']

column= pn.Column("## Option Chain", interact (option_chain, symbol = march_symbol), "## Volatility Smiles", row)

# column= pn.Column("## Option Chain", march_demo_df, "## Volatility Smiles", row)
#column= pn.Column("## Option Chain", march_demo_df1, "## Volatility Smiles", row)
```

```python
row1 = pn.Row(interact(plot_smile3, symbol = april_symbol),interact(plot_smile4, symbol = april_symbol))

# april_demo_df = april_maturities.loc[april_maturities['symbol']=='AAPL']
# april_demo_df1 = april_maturities.loc[april_maturities['symbol']=='SPY']
column1= pn.Column("## Option Chain", interact (option_chain1, symbol = april_symbol), "## Volatility Smiles", row1)

# column1= pn.Column("## Option Chain", april_demo_df, "## Volatility Smiles", row1)
# column1= pn.Column("## Option Chain", april_demo_df1, "## Volatility Smiles", row1)
```
